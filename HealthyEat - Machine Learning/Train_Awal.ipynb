```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.utils import img_to_array , load_img
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dropout, Flatten, Dense
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
```

# Data Preprocessing

```python
train_path = '/content/Training'
test_path = '/content/Test'
val_path = '/content/Validation'
```

```python
img_width = 180
img_height = 180
```

```python
train_path = tf.keras.utils.image_dataset_from_directory(
    train_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32)
```

```python
data_cat1 = train_path.class_names
```

```python
test_path = tf.keras.utils.image_dataset_from_directory(
    test_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32)
```

```python
data_cat2 = test_path.class_names
```

```python
val_path = tf.keras.utils.image_dataset_from_directory(
    val_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32)
```

```python
data_cat3 = val_path.class_names
```

```python
plt.figure(figsize=(7,7))
for image, labels in train_path.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_cat1[labels[i]])
        plt.axis('off')
```

```python
plt.figure(figsize=(7,7))
for image, labels in test_path.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_cat2[labels[i]])
        plt.axis('off')
```

```python
plt.figure(figsize=(7,7))
for image, labels in val_path.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_cat3[labels[i]])
        plt.axis('off')
```

# Create Model

```python
model = Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3, padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128),
    layers.Dense(len(data_cat1))

])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
#Train the Model
epochs_size = 150
history = model.fit(train_path, validation_data=val_path, epochs=epochs_size)
```

```python
epochs_range = range(epochs_size)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(epochs_range,history.history['accuracy'],label = 'Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'],label = 'Validation Accuracy')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range,history.history['loss'],label = 'Training Loss')
plt.plot(epochs_range, history.history['val_loss'],label = 'Validation Loss')
plt.title('Loss')
```

```python
# Simpan model ke file .h5
model.save('model.h5')
```
